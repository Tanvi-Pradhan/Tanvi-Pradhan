## Derived Metrics
Now, Creating bins for continous variable categories column 'AMT_INCOME_TOTAL' , 'AMT_CREDIT', 'AGE', 'YEARS_EMPLOYED'

```python
##Binning AMT_INCOME_TOTAL as INCOME_CAT

df['INCOME_CAT']=pd.cut(
 df['AMT_INCOME_TOTAL'],
 bins = [0,50000,100000,150000,200000,250000,300000,350000,400000,450000,500000,np.inf],
 labels = ["'<50K'","'50-100K'","'100-150K'","'150-200K'","'200-250K'","'250-300K'","'300-350K'","'350-400K'","'400-450K'","'450-500K'","'>500K'"]
)
```

```python
##Binning AMT_CREDIT as CREDIT_CAT

df['CREDIT_CAT']=pd.cut(
 df['AMT_CREDIT'],
 bins = [0,100000,200000,300000,400000,500000,600000,700000,800000,900000,1000000,np.inf],
 labels = ["'<100K'","'100-200K'","'200-300K'","'300-400K'","'400-500K'","'500-600K'","'600-700K'","'700-800K'","'800-900K'","'900-1000K'","'>1000K'"]
)
```

```python
#Creating a age Column and binnng it as AGE_CAT

df['AGE']=abs(df['DAYS_BIRTH']//365.25)
df['AGE_CAT']=pd.cut(
 df['AGE'],
 bins=[0, 30, 40, 50, 60, np.inf],
 labels = ["'<30'", "'30-40'", "'40-50'", "'50-60'", "'>60'"]
)
```

```python
#Creating a age Column and binnng it as YEARS_EMPLOYED
df['YEARS_EMPLOYED']=abs(df['DAYS_EMPLOYED'])//365.25
df['YEARS_EMPLOYED']=pd.cut(
 df['YEARS_EMPLOYED'],
 bins = [0,10,15,20,25,30,35,40,45,50,np.inf],
 labels = ["'<10'","'10-15'","'15-20'","'20-25'","'25-30'","'30-35'","'35-40'","'40-45'","'45-50'","'>50'"]
)
```
